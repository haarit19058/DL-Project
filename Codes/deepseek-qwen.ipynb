````python
# Use unsloth quantized model
from unsloth import FastModel
model, tokenizer = FastModel.from_pretrained(
    model_name="unsloth/Qwen3-0.6B-Base-unsloth-bnb-4bit",
    max_seq_length=2048,
)

# use unsloth for lora
model = FastVisionModel.get_peft_model(
    model,
    finetune_vision_layers     = False, # False if not finetuning vision layers
    finetune_language_layers   = True,  # False if not finetuning language layers
    finetune_attention_modules = True,  # False if not finetuning attention layers
    finetune_mlp_modules       = True,  # False if not finetuning MLP layers

    r = 16,           # The larger, the higher the accuracy, but might overfit
    lora_alpha = 16,  # Recommended alpha == r at least
    lora_dropout = 0,
    bias = "none",
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
    use_gradient_checkpointing = "unsloth", # Reduces memory usage
    # target_modules = "all-linear", # Optional now! Can specify a list if needed
)



# dataset is codeforces cots given by deepseek
# columns of the dataset 
# Index(['id', 'aliases', 'contest_id', 'contest_name', 'contest_type',
#    'contest_start', 'contest_start_year', 'index', 'time_limit',
#    'memory_limit', 'title', 'description', 'input_format', 'output_format',
#    'examples', 'note', 'editorial', 'prompt', 'generation',
#    'finish_reason', 'api_metadata', 'interaction_format', 'messages'],
#   dtype='str')
# The prompt column contains are premade prompt by deep seek, we need to give stuff like following
def getdata(row):
    return {
        "question_level":row['id'].split('/')[-1],
        "contest_name":row['contest_name'],
        "contest_type":row['contest_type'],
        "time_limit":str(row['time_limit'])+' seconds',
        'memory_limit':str(row['memory_limit'])+' MB',
        'prompt':row['prompt'] # prepared prompt that we need to give toe the model
    }

# I need to train my model on only question A from the dataset and not all the questions availableso filter the A question
DATASET_NAME = "open-r1/codeforces-cots"
dataset = load_dataset(DATASET_NAME, DATASET_CONFIG, split="train[:1000]") # use a subset of dataset for now



# now once we have load the model and the dataset we move on to make a proper reward function for GRPO
# we need to give rewards according to the formatting, compilation of code / working successfully without any errors and the number of test cases passed
#  for this poriton here is the example code snippet
# I need you to deep search the web and use the same rewards as deepseek used to train their coding models
def format_reward_func(completions, **kwargs):
    rewards = []
    for c in completions:
        # FIX: Added triple backticks to properly match Markdown blocks
        has_think = bool(re.search(r"<think>.*?</think>", c, re.DOTALL))
        has_code = bool(re.search(r"```(python|cpp)\n.*?```", c, re.DOTALL))

        # FIX: Fixed indentation so it appends for every completion
        rewards.append(0.2 if (has_think and has_code) else 0.0)
    return rewards

def extract_code(completion):
    # FIX: Added triple backticks
    match = re.search(r"```(python|cpp)\n(.*?)```", completion, re.DOTALL)
    if match:
        return match.group(1), match.group(2)
    return None, None

def run_code(code, lang, inp):
    with tempfile.TemporaryDirectory(prefix=f"rank{local_rank}_") as tmp:
        if lang == "cpp":
            cpp = os.path.join(tmp, "sol.cpp")
            exe = os.path.join(tmp, "sol")
            open(cpp, "w").write(code)

            if subprocess.run(["g++", "-O2", cpp, "-o", exe]).returncode != 0:
                return None, "ce"
            cmd = [exe]
        else:
            py = os.path.join(tmp, "sol.py")
            open(py, "w").write(code)
            cmd = [sys.executable, py]

    try:
        res = subprocess.run(cmd, input=inp, text=True, capture_output=True, timeout=2)
        if res.returncode != 0:
            return None, "re"
        return res.stdout.strip(), "ok"
    except subprocess.TimeoutExpired:
        return None, "tle"
    except Exception as e:
        return None, f"error: {str(e)}"

def cp_reward(completions, input_tests, output_tests, **kwargs):
    rewards = []
    for comp, ins, outs in zip(completions, input_tests, output_tests):
        lang, code = extract_code(comp)

        if not code:
            rewards.append(-0.5)
            continue

        score = 0
        total = len(ins)

        for i, o in zip(ins, outs):
            out, status = run_code(code, lang, i)
            if status == "ok" and out == o.strip():
                score += 1

        rewards.append(score / total if total else -0.5)

    return rewards


# Once all the components are ready i need you to use the grpo trainer to train the model
from datasets import load_dataset
from trl import GRPOConfig, GRPOTrainer
from trl import GRPOConfig, GRPOTrainer
training_args = GRPOConfig(
    learning_rate = 5e-6,
    adam_beta1 = 0.9,
    adam_beta2 = 0.99,
    weight_decay = 0.1,
    warmup_ratio = 0.1,
    lr_scheduler_type = "cosine",
    optim = "adamw_8bit",
    logging_steps = 1,
    log_completions = False,
    per_device_train_batch_size = 1,
    gradient_accumulation_steps = 1, # Increase to 4 for smoother training
    num_generations = 2, # Decrease if out of memory
    max_prompt_length = 1024,
    max_completion_length = 1024,
    # num_train_epochs = 0.5, # Set to 1 for a full training run
    max_steps = 60,
    save_steps = 60,
    max_grad_norm = 0.1,
    report_to = "none", # Can use Weights & Biases
    output_dir = "outputs",

    # Below enables GSPO:
    importance_sampling_level = "sequence",
    mask_truncated_completions = False,
    loss_type = "dr_grpo",
)
trainer = GRPOTrainer(
    model = model,
    args = training_args,
    # Pass the processor to handle multimodal inputs
    processing_class = tokenizer,
    reward_funcs = [
        formatting_reward_func,
        correctness_reward_func,
    ],
    train_dataset = train_dataset,
)

trainer.train()

# use the grpo to train our model 
# during training i need you to do detailed loggings to see what happening at each iteration
# after each epoch i need you to save the model params (under same fille name and not separate weight pkl for each epoch) so that it is not lost in between and if the code crashes i can revive the code using those weights




# once the training is complete for manual verification i need you to take a question and use the model to generate its code and run the reward metrices earlier and print them
# the aim of htis activity is to understand the final output and formatting that the model is generating and how many test cases are passing


````
